### 교차검증

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target
df_clf = DecisionTreeClassifier(random_state=156)

# k= 5
kfold = KFold(n_splits=5) # features의 4/5는 train, 1/5은 test data가 됨
cv_accuracy = [] # 매 학습 때마다의 성능 결과치
```

```python
n_iter = 0

# split 5로 설정해놨기 때문에 이 loop를 5번 돌게 됨
# kfold.split(features)는 index를 반환
for train_index, test_index in kfold.split(features):
    print(train_index, test_index)
```

```python
for train_index, test_index in kfold.split(features):
    X_train, X_test = features[train_index], features[test_index]
    y_train, y_test = label[train_index], label[test_index]
    
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    n_iter += 1
    
    accuracy = np.round(accuracy_score(y_test, pred), 4)
    cv_accuracy.append(accuracy)
    
print(cv_accuracy)
print(np.mean(cv_accuracy))
```

Stratified K 폴드

```python
import pandas as pd
iris = load_iris()

iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names) # value_counts 사용하기 위해 df로 변경
iris_df['label'] = iris.target

iris_df['label'].value_counts()
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=3)
n_iter = 0

for train_index, test_index in skf.split(iris_df, iris_df['label']): # y값을 넣어줘서 분포가 고루고루 들어가게끔
    n_iter += 1
    label_train = iris_df['label'].iloc[train_index]
    label_test = iris_df['label'].iloc[test_index]
    
    print("{0} 번째 학습".format(n_iter))
    print('학습 레이블 데이터 분포:\n', label_train.value_counts())
    print('검증 레이블 데이터 분포:\n', label_test.value_counts())
```

cross_val_score()

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, cross_validate
from sklearn.datasets import load_iris
import numpy as np

iris_data = load_iris()
dt_clf = DecisionTreeClassifier(random_state=156)

data = iris_data.data
label = iris_data.target

scores = cross_val_score(dt_clf, data, label, scoring='accuracy', cv=3)
print('교차 검증별 정확도:', scores)
print('평균 검증 정확도:', np.mean(scores))
```

GridSearchCV

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=121)

dtree = DecisionTreeClassifier()

parameters = {'max_depth': [1,2,3], 'min_samples_split':[2,3]}
```

```python
import pandas as pd

# refit=True -> 최적의 하이퍼파라미터를 찾으면 그것으로 모델을 교체
grid_dtree = GridSearchCV(dtree, param_grid=parameters, cv=3, refit=True, return_train_score=True)

# parameter 조합 6 * cv 3 = 18번 수행
grid_dtree.fit(X_train, y_train)

scores_df = pd.DataFrame(grid_dtree.cv_results_)
scores_df[['params', 'mean_test_score', 'rank_test_score', 'split0_test_score', 'split1_test_score', 'split2_test_score']]
```

```python
print("최적 파라미터:", grid_dtree.best_params_)

# refit=True여서 학습이 되어있음
pred = grid_dtree.predict(X_test)
print(accuracy_score(y_test, pred))
```

```python
estimator = grid_dtree.best_estimator_

pred = estimator.predict(X_test)
print(accuracy_score(y_test, pred))
```

### 타이타닉 생존자 예측

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

titanic_df = pd.read_csv('./titanic_train.csv')
titanic_df.head(3)
```

```python
print("## train 데이터 정보 ##")
print(titanic_df.info())
```

```python
# 결측치 처리
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
titanic_df['Cabin'].fillna('N', inplace=True)
titanic_df['Embarked'].fillna('N', inplace=True)

print(titanic_df.isnull().sum())
```

```python
titanic_df['Sex'].value_counts()
titanic_df['Cabin'].value_counts()
titanic_df['Embarked'].value_counts()
```

```python
titanic_df['Cabin'] = titanic_df['Cabin'].str[:1]
titanic_df['Cabin'].value_counts()
```

```python
titanic_df.groupby(['Sex', 'Survived']).count()
titanic_df.groupby(['Sex', 'Survived'])['Survived'].count()
```

```python
# 성별과 생존여부의 관계 살펴보기
sns.barplot(x='Sex', y='Survived', data=titanic_df)
```

```python
# 성별, 선실등급과 생존여부의 관계
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=titanic_df)
```

```python
# 나이와 생존여부 관계
def get_category(age):
    cat = ''
    if age <= -1: cat = 'Unknown'
    elif age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenager'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    
    return cat

# 막대그래프의 크기 figure를 더 크게 설정 
plt.figure(figsize=(10,6))

group_names = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student', 'Young Adult', 'Adult', 'Elderly']

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x: get_category(x))
sns.barplot(x='Age_cat', y='Survived', hue='Sex', data=titanic_df, order=group_names)
titanic_df.drop('Age_cat', axis=1, inplace=True)
```

```python
from sklearn import preprocessing

def encode_features(dataDF):
    features = ['Cabin', 'Sex', 'Embarked']
    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature] = le.transform(dataDF[feature])
        
    return dataDF

titanic_df = encode_features(titanic_df)
titanic_df
```
